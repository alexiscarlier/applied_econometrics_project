```python
def intersection(lst1, lst2): 
    return list(set(lst1) & set(lst2)) 
```

```python
clean_post_78[clean_post_78['home_id'] == '162311']
```

```python
len(set(list(clean_before_78['home_id'])))
```

```python
all_houses.head()
```

## Loading the electricity and natural gas prices to add to dfs

```python
electricty = pd.read_excel('avgprice_annual_electricity.xlsx')
```

```python
natural_gas = pd.read_excel('natural_gas_price.xls')
```

Reducing the electricty data to just include that from Florida

```python
electricity = electricty[electricty['State'] == 'FL']
```

## Note that the unit for the electricity measurements is cents/killowatthour

```python
electricity.head()
```

#### Deleting all the electricity stuff I don't need

```python
electricity = electricity[electricity['Industry Sector Category'] == 'Total Electric Industry']
electricity.head()

cleaned_electricity = electricity.drop(['State', 'Industry Sector Category', 'Industrial', 'Commercial', 'Transportation', 'Other', 'Total'], axis=1)
```

### Idea: merge the dataframes on year. First, change the year column so it's the same in both, then merge

```python
cleaned_electricity = cleaned_electricity.rename(columns={'Year': 'year'})
```

```python
all_houses = pd.merge(all_houses, cleaned_electricity, on='year', how='outer')
```

```python
all_houses = all_houses.rename(columns={'Residential': 'avg electricity price cents/kwh'})
```

```python
all_houses
```

#### Same process now for natural gas

```python
natural_gas.head()
```

```python
natural_gas = natural_gas.rename(columns={'Florida Price of Natural Gas Delivered to Residential Consumers (Dollars per Thousand Cubic Feet)': 'avg natural gas price $/thousand cubic feet'})
```

```python
natural_gas['Date'] = natural_gas['Date'].apply(lambda x: x.year)
```

```python
natural_gas = natural_gas.rename(columns={'Date': 'year'})
```

```python
all_houses = pd.merge(all_houses, natural_gas, on='year', how='outer')
```
